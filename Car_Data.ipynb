Preprocessing of data has to be done before giving the data for more detailed analysis.

```python
#Splitting the Used Car Model into Year and Car Model but took 0 = Year and 1 = Car Model
df[['0','1']]= df['Used Car Model'].str.split(' ',n=1,expand=True)
```

```python
#Renaming the 0 and 1 columns into Year and Car Model respectively
df = df.rename(columns={'0' : 'Year' ,'1' : 'Car Model'})
df
```

```python
#Dropping the main column which I used for splitting
df = df.drop(['Used Car Model'],axis=1)
df
```
